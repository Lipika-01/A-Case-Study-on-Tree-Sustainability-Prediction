```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/tree_sustainability_data.xlsx")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.drop(columns=["observation_date"],inplace=True)
df.drop(columns=["observation_time"],inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"tree_height":"Tree-Height"},inplace=True)
df.rename(columns={"tree_diameter":"Tree-Diameter"},inplace=True)
df.rename(columns={"age":"Age"},inplace=True)
df.rename(columns={"soil_ph":"PH_of_Soil"},inplace=True)
df.rename(columns={"soil_type":"Type_of_Soil"},inplace=True)
df.rename(columns={"rainfall":"Rainfall"},inplace=True)
df.rename(columns={"temperature":"Temperature"},inplace=True)
df.rename(columns={"sun_hours":"Sunlight"},inplace=True)
df.rename(columns={"slope":"Slope"},inplace=True)
df.rename(columns={"elevation":"Elevation"},inplace=True)
df.rename(columns={"land_use":"Land_Use"},inplace=True)
df.rename(columns={"proximity_water":"Proximity_Water"},inplace=True)
df.rename(columns={"competition_index":"Competition_Index"},inplace=True)
df.rename(columns={"disease_index":"Disease_Index"},inplace=True)
df.rename(columns={"pollution_level":"Pollution_Level"},inplace=True)
df.rename(columns={"sustainable":"Is_Sustainable?"},inplace=True)
```

```python
df.head()
```

```python
sns.histplot(df["Rainfall"])
plt.plot()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Type_of_Soil"] = le.fit_transform(df["Type_of_Soil"])
df["Land_Use"] = le.fit_transform(df["Land_Use"])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
df["Tree-Height"] = ss.fit_transform(df[["Tree-Height"]])
df["Tree-Diameter"] = ss.fit_transform(df[["Tree-Diameter"]])
df["Age"] = ss.fit_transform(df[["Age"]])
df["Temperature"] = ss.fit_transform(df[["Temperature"]])
df["Sunlight"] = ss.fit_transform(df[["Sunlight"]])
df["Slope"] = ss.fit_transform(df[["Slope"]])
df["Elevation"] = ss.fit_transform(df[["Elevation"]])
df["Rainfall"] = ss.fit_transform(df[["Rainfall"]])
```

```python
df.head()
```

```python
x = df.drop(columns=["Is_Sustainable?"])
y = df["Is_Sustainable?"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
ypred = model.predict(x_test)
```

```python
print(ypred)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,ypred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,ypred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,ypred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
ypred = model.predict(x_test)
```
